```python

#print(h)

rhoc = 2.7
rhom = 3.3
Hr=[]
for i in range(len(h)):
    H = ((h[i]*rhoc)/(rhom-rhoc))
    Hr.append(H)
    
Gr = []
for j in range(len(Hr)):
    G = 0.04193*rhoc*Hr[j]
    Gr.append(G)
    
    
Gi = Gb - Gr

Overc=[]
Underc=[]
completec=[]

Underc = Gi[Gi > 0].tolist()
Overc = Gi[Gi < 0].tolist()
completec = Gi[Gi == 0].tolist()


Underc = Gi[Gi > 0]
Overc = Gi[Gi < 0]
completec = Gi[Gi == 0]

latu = lat[Gi > 0]
lato = lat[Gi < 0]
latc = lat[Gi == 0]

longu = long[Gi > 0]
longo = long[Gi < 0]
longc = long[Gi == 0]


# Plotting the data
plt.figure(figsize=(10, 6))

# Scatter plots for each dataset
plt.scatter(longo, lato, c=Overc, cmap='Reds', label='Dataset 1', s=50, alpha=0.8)
plt.scatter(longu, latu, c=Underc, cmap='Greens', label='Dataset 2', s=50, alpha=0.8)
plt.scatter(longc, latc, c=completec, cmap='Blues', label='Dataset 3', s=50, alpha=0.8)


plt.xlim(60,105)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Gravity Anomaly Data')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import folium

# Load the Excel file into a pandas DataFrame
excel_file = 'isostasy_anomaly1.xlsx'
df = pd.read_excel(excel_file)

# Create a map centered around a specific location
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
mymap = folium.Map(location=map_center, zoom_start=5)

# Add markers for each data point
for index, row in df.iterrows():
    location = [row['Latitude'], row['Longitude']]
    gravity_value = row['Isostasy_anomaly']
    popup_text = f"Latitude: {row['Latitude']}, Longitude: {row['Longitude']}, Gravity: {gravity_value}"
    folium.Marker(location=location, popup=popup_text).add_to(mymap)

# # Save the map to an HTML file
# mymap.save('gravity_map.html')
mymap
```
